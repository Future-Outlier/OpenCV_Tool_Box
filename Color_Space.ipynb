## HSV Color Space
### RGB TO HSV Formuala


### H  S  V  
### H ""S ""V ""


```python
img = cv2.imread("images/colorful_forest.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
img_hstack = np.hstack((img, img_hsv)) 
plt.imshow(img_hstack)
plt.show()
```

##  V 0

```python
H, S, V = cv2.split(img_hsv)
V_new = np.zeros(V.shape)
new_hsv = np.copy(img)
new_hsv[:, :, 0], new_hsv[:, :, 1], new_hsv[:, :, 2] = H, S, V_new
img_hstack = np.hstack((new_hsv, cv2.cvtColor(new_hsv, cv2.COLOR_HSV2RGB))) 
plt.imshow(img_hstack)
plt.show()
```

# Hue , Saturation , Value  

```python
H, S, V = cv2.split(img_hsv)
print(np.mean(H)), print(np.mean(S)), print(np.mean(V))
hue = np.copy(H)
saturation = np.copy(S)
value = np.copy(V)
hue.fill(200), saturation.fill(255), value.fill(255)
hue_200_img = cv2.cvtColor(cv2.merge([hue, S, V]), cv2.COLOR_HSV2RGB)
saturation_img = cv2.cvtColor(cv2.merge([H, saturation, V]), cv2.COLOR_HSV2RGB)
value_img = cv2.cvtColor(cv2.merge([H, S, value]), cv2.COLOR_HSV2RGB)
hue.fill(50)
hue_50_img =  cv2.cvtColor(cv2.merge([hue, S, V]), cv2.COLOR_HSV2RGB)
hue.fill(100)
hue_100_img =  cv2.cvtColor(cv2.merge([hue, S, V]), cv2.COLOR_HSV2RGB)
hue.fill(255)
hue_255_img =  cv2.cvtColor(cv2.merge([hue, S, V]), cv2.COLOR_HSV2RGB)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Orginal Image")
plt.subplot(2,2,2)
plt.imshow(hue_200_img)
plt.title("Larger Hue Image")# try 50 255 few
plt.subplot(2,2,3)
plt.imshow(saturation_img)
plt.title("Larger Saturation Image")
plt.subplot(2,2,4)
plt.imshow(value_img)
plt.title("Larger Value Image")
```

# Alpha  
## OpenCVBGRBGRA
### :APICODEXD

```python
image = cv2.imread('images/self_image.jpg')
cv2.imshow("The Image", image)                  # BGR

bgra_image = cv2.cvtColor(image, cv2.COLOR_BGR2BGRA)
b, g, r, a = cv2.split(bgra_image)
print("Aalpha")
print(a)

a[:,:] = 32                                     # alpha
a32_image = cv2.merge([b, g, r, a])             # alpha=32
cv2.imshow("The a32 Image", a32_image)          # alpha=32

a.fill(128)                                     # alpha
a128_image = cv2.merge([b, g, r, a])            # alpha=128
cv2.imshow("The a128 Image", a128_image)        # alpha=128

cv2.waitKey(0)
cv2.destroyAllWindows()
      
cv2.imwrite('images/a32.png', a32_image)               # alpha=32
cv2.imwrite('images/a128.png', a128_image)             # alpha=128   
```
