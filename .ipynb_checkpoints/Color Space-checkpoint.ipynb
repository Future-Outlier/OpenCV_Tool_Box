```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("self_image.jpg") # default is BGR colimgor space
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
BGR_and_RGB_img = np.hstack((img, RGB_img))
plt.imshow(BGR_and_RGB_img)
plt.show()
```

## Gray = 0.2989 * R + 0.5870 * G + 0.1140 * B

## HSV Color Space
### RGB TO HSV Formuala


### H 是色相 S 是飽和度 V 是 明度
### H 是圓柱的角度"色相"，S 是軸的角度代表"飽和度"，V 是圓柱的高度，代表"明度"


```python
img = cv2.imread("colorful_forest.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
img_hstack = np.hstack((img, img_hsv)) 
plt.imshow(img_hstack)
plt.show()
```

## 當 V 明度設定為0時，影像為全黑

```python
H, S, V = cv2.split(img_hsv)
V_new = np.zeros(V.shape)
new_hsv = np.copy(img)
new_hsv[:, :, 0], new_hsv[:, :, 1], new_hsv[:, :, 2] = H, S, V_new
img_hstack = np.hstack((new_hsv, cv2.cvtColor(new_hsv, cv2.COLOR_HSV2RGB))) 
plt.imshow(img_hstack)
plt.show()
```
