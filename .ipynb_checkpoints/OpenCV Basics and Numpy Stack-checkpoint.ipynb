```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("self_image.jpg")
```

```python
cv2.namedWindow("Test picture", cv2.WINDOW_NORMAL)
cv2.imshow("Test picture", img)
ret_value = cv2.waitKey(0)
if ret_value == ord('Q') or ret_value == ord('q'):
    cv2.destroyWindow("Test picture")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img)
plt.show()
```

```python
print(f"shape = {img.shape}")
print(f"size = {img.size}")
print(f"dtype = {img.dtype}")
print("Shape Dimension , height, width, channel")
print("Shape Dimension , y, x, channel")
```

## numpy axis可當作多維陣列的索引

```python
np.vstack # 垂直合併陣列
np.hstack # 水平合併陣列
```

```python
two_vertical_img = np.vstack((img, img))
plt.imshow(two_vertical_img)
plt.show()
two_horizontal_img = np.hstack((img, img))
plt.imshow(two_horizontal_img)
plt.show()
four_square_img = np.hstack((two_vertical_img, two_vertical_img))
plt.imshow(four_square_img)
plt.show()
```
